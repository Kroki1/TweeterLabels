```python
df['keyword'].value_counts()
```

```python
table = pd.pivot_table(df, index='keyword', columns='target',values='id', aggfunc='count')
table
```

```python
table.reset_index(inplace=True)
```

```python
table[0].fillna(0, inplace=True)
table[1].fillna(0, inplace=True) 
table['sum'] = (table[0] + table[1])
table['dis%'] = table[1]/(table[0] + table[1])
table
```

```python
table.sort_values(by=['dis%'], ascending=False, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(200,50))
ax = sns.barplot(data=table, x="keyword", y="dis%")
```

```python
table.to_csv('./Data/table.csv', encoding='utf-8')
```
